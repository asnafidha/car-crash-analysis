```python
ccd.drop_duplicates(inplace=True)
```

```python

ccd['Weekend?'].fillna(ccd['Weekend?'].mode()[0], inplace=True)

ccd['Hour'].fillna(ccd['Hour'].median(), inplace=True)

ccd['Collision Type'].fillna(ccd['Collision Type'].mode()[0], inplace=True)

ccd['Primary Factor'].fillna(ccd['Primary Factor'].mode()[0], inplace=True)

# Drop rows with missing values in geographical columns
ccd.dropna(subset=['Reported_Location', 'Latitude', 'Longitude'], inplace=True)
```

## DATA VISUALISATION AND ANALYSIS

1.**DISTRIBUTION OF CAR CRASHES OVER TIME**

```python
plt.figure(figsize=(12, 6))
ccd['timestamp'] = pd.to_datetime(ccd[['Year', 'Month', 'Day', 'Hour']])
monthly_crash_count = ccd.resample('M', on='timestamp').size()
monthly_crash_count.plot(marker='o', linestyle='-', color='b')
plt.title('Monthly Distribution of Car Crashes')
plt.xlabel('Time')
plt.ylabel('Number of Car Crashes')
plt.grid(True)
plt.show()
```

*ACCIDENTS WAS MOSTLY OCCURED IN YEAR OF 2003,2004,2008,2009 AND 2015


2 **DISTRIBUTION OF COLLISION TYPE**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Collision Type', data=ccd, palette='viridis')
plt.title('Distribution of Collision Types')
plt.xlabel('Collision Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

*ACCIDENTS MOSTLY WERE 2 CARS ACCIDENT AND IT IS ABOVE 35000.

*CYCLIST AND PEDESTRIAN TYPES WAS LOWEST

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Injury Type', data=ccd, palette='muted')
plt.title('Distribution of Injury Types')
plt.xlabel('Injury Type')
plt.ylabel('Count')
plt.show()
```

3 **WEEKDAY VS WEEKEND**

```python
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
sb.countplot(x=ccd['Weekend?'], palette='Accent')
plt.ylabel('Frequency')
plt.xticks(rotation=65)
plt.title('Weekday Vs Weekend')
plt.xlabel('Day Type (0: Weekday, 1: Weekend)')
plt.show()
```

*crashes were mostly occured in weekdays

4 **GEOGRAPHIC DISTRIBUTION OF CAR CRASHES**

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', data=ccd, hue='Collision Type', palette='coolwarm', alpha=0.5)
plt.title('Geographic Distribution of Car Crashes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Collision Type')
plt.show()
```

5 **MOST COMMON REASON FOR ACCIDENTS**

```python
ccd['Primary Factor'].nunique()
```

```python
cdddata=ccd['Primary Factor'].value_counts().head(20)
```

```python
fig=plt.figure(figsize=(10,10))
axis=fig.add_axes([1,1,1,1])
sb.swarmplot(data=cdddata,y=cdddata.index,x=cdddata,ax=axis)
for i,j in enumerate(cdddata):
    axis.text(j,i,j)
plt.xlabel('Occurrence')
plt.title('Most Common Reason For Accident')
```

6 **MOST FREQUENT LOCATION OF ACCIDENTS**

```python
location_counts = ccd['Reported_Location'].value_counts()
top_locations = location_counts.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.index, y=top_locations.values, palette='viridis')
plt.title('Top 10 Most Frequent Locations of Accidents')
plt.xlabel('Location')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')
plt.show()
```

7 **PAIRPLOT**

```python
 # Pairplot for selected numerical features
numerical_features = ['Year', 'Month', 'Day', 'Hour']
sns.pairplot(ccd[numerical_features], height=2.5)
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```
